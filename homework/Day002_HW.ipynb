註: with 開啟檔案時，會將開啟的檔案放在 csvfile變數中，但是這個csvfile只有在這個 with 的範圍內可以使用，

```python
import csv

# 開啟 CSV 檔案
with open('./data/example.csv', newline='',encoding='UTF-8') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    # 以迴圈輸出每一列，若此二行未拿掉，下面的example_data=[]，變空的
    #for row in rows:   
    #    print(row)

    example_data=list(rows)
```

```python
# 1. 取出班次一的每一個時間
for x in range(len(example_data)): 
    print(example_data[x][5])
```

* 重點是在如何把新資料加入list中
* http://smilejay.com/2013/02/add-items-to-a-list-in-python/
* car1.append=example_data[x][5] --> 錯誤寫法 
* 會出現'list' object attribute 'append' is read-only

```python
# 2. 將班次一的每一個時間用一種資料型態保存
car1=[] #先建一個空的list
for x in range(len(example_data)): 
    car1.append(example_data[x][5])
print(car1)
```

```python
# 3. 將班次一到五與其所有時間用一種資料型態個別保存
car1to5=[] #先建一個空的list
for x in range(len(example_data)): 
    car1to5.append(example_data[x][5:10])
print(car1to5)
```
